```python
long_window = 21
short_window = 7

def SMAwindowSet(df):
    df['Signal'] = 0.0

    df['SMAshort'] = df['close'].rolling(window=short_window).mean()
    
    df['SMAlong'] = df['close'].rolling(window=long_window).mean()
    
    df['Signal'][short_window:] = np.where(
    df['SMAshort'][short_window:] > df['SMAlong'][short_window:], 1.0, 0.0)
    
    df['Entry/Exit'] = df['Signal'].diff()
    return df
   

SMAwindowSet(BTCM_df)
    
    
    
```
