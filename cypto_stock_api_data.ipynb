```python
# Be sure to set the values from the DataFrame to a float by calling the float function 
from pylab import mpl, plt
plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'

WATT_data = (portfolio_prices_df["WATT"])

FINV_data = (portfolio_prices_df["FINV"])

BTCM_data = (portfolio_prices_df["BTCM"])
```

```python
BTCM_df = pd.DataFrame(BTCM_data.drop(columns= ['open', 'high', 'low', 'volume']))

BTCM_df['SMA_short'] = BTCM_df['close'].rolling(7).mean()
BTCM_df['SMA_long'] = BTCM_df['close'].rolling(15).mean()

#display(BTCM_df.tail(25))

#BTCM_df.plot(title='BTCM | 20 & 100 days SMAs', figsize=(15,7))

import numpy as np
```

```python
BTCM_df['position'] = np.where(BTCM_df['SMA_short'] > BTCM_df['SMA_long'],
                             1, -1)

BTCM_df.dropna(inplace=True)
```

```python
BTCM_df['position'].plot(ylim=[-1.1, 1.1],
                       title="Market Positioning",
                       figsize=(15,7))
```

```python
BTCM_df['returns'] = np.log(BTCM_df['close'] / BTCM_df['close'].shift(1))

```

```python
BTCM_df['returns'].hist(bins=25, figsize=(15,7))
```

```python
BTCM_df['strategy'] = BTCM_df['position'].shift(1) * BTCM_df['returns']
```

```python
BTCM_df[['returns', 'strategy']].sum()
```

```python
BTCM_df[['returns', 'strategy']].sum().apply(np.exp)
```

```python
BTCM_df[['returns', 'strategy']].cumsum().apply(np.exp).plot(figsize=(15,7))
```

```python
BTCM_df[['returns', 'strategy']].mean() * 252
```

```python
np.exp(BTCM_df[['returns', 'strategy']].mean() * 252) -1
```

```python
BTCM_df[['returns', 'strategy']].std() * 252 ** 0.5
```

```python
(BTCM_df[['returns', 'strategy']].apply(np.exp) - 1).std() * 252 ** 0.5
```

```python
BTCM_df['cumret'] = BTCM_df['strategy'].cumsum().apply(np.exp) 
```

```python
BTCM_df['cummax'] = BTCM_df['cumret'].cummax()
```

```python
BTCM_df[['cumret', 'cummax']].dropna().plot(figsize=(15,7))
```
